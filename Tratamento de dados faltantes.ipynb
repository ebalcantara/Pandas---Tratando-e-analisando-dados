# Alguns problemas identificados

    Colunas que não podem ter valor nulo ou zerados : ['Valor']

    Se o tipo de imóvel for apartamento os dados da coluna 'Condominio' não podem ser nulos ou zero
    

## Tratamento dos dados na coluna 'Valor'

```python
dados.isnull() #booleano de informações nulas
```

```python
dados.notnull() #booleano de informações não nulas
```

```python
dados.info() #informações gerais do DataFrame
```

```python
dados[dados['Valor'].isnull()] #DataFrame apenas com a coluna 'valor' com valores nulos
```

```python
# Eliminação do DataFrame os valores nulos da coluna 'Valor'
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
```

```python
dados[dados['Valor'].isnull()] 
```

## Tratamento dos dados da coluna 'Condominio'

```python
dados[dados['Condominio'].isnull()].shape[0] #Identificando a quantidade de imóveis com o condomínio nulo
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull()) #Selecionando os apartamentos que estão com o condomínio com valor nulo
```

```python
# Eliminação do DataFrame os valores nulos da seleção acima
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0] #Vendo novamente os valores nulos na coluna 'Condomínio'
```

## Tratamento dos dados que estão com o condomínio e IPTU nulo.
    Esse dados não serão eliminados por conterem informações importantes
    para serem considerados, vou considerar que os valores que são nulos agora serão preenchidos com valor 0

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0}) #Substituindo os valores nulos por 0
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

```python
dados.info()
```

## Substituindo os dados do arquivo original pelos dados tratados.
