## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

#### https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats

```python
dados['Valor'].mean()
```

```python
Bairros = list(dict.fromkeys(list(dados.Bairro)))
selecao = dados['Bairro'].isin(Bairros)
dados = dados[selecao]
dados
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, dados.Valor.mean().round(2)))
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

```python
grupo_bairro['Valor'].describe().round(2).head(50)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'}).head(20)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
