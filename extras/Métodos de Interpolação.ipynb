```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.53, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
```

```python
s
```

```python
s.fillna(0)
```

```python
s.fillna(method = 'ffill') #metodo de interpolação que preenche os valores nulos com o valor anterior na série
```

```python
s.fillna(method = 'bfill') #metodo de interpolação que preenche os valores nulos com o valor seguinte na série
```

```python
s.fillna(method = 'ffill', limit = 1) #metodo de interpolação que preenche os valores nulos com o valor anterior na série porém não repete mais que 1 vez
```
